## üß± Model: Pru≈æina s tlumen√≠m

Studujeme syst√©m:

- Hmotnost: \(m\)
- Tuhost pru≈æiny: \(k\)
- Tlumen√≠: \(b\)
- Pozice: \(x\)

Budeme odvozovat stejn√Ω stavov√Ω model pomoc√≠ t≈ô√≠ metod:

1. Newtonovsk√© mechaniky  
2. Lagrangeovy mechaniky  
3. Hamiltonovy mechaniky  

---

## 1Ô∏è‚É£ Newtonovsk√° mechanika (s√≠ly)

Pou≈æijeme 2. Newton≈Øv z√°kon:

$$
F = m a = m \ddot{x}
$$

S√≠ly:
- Pru≈æina: \(F_k = -k x\)  
- Tlumen√≠: \(F_b = -b \dot{x}\)

Rovnice pohybu:

$$
m \ddot{x} = -k x - b \dot{x} \Rightarrow \ddot{x} + \frac{b}{m} \dot{x} + \frac{k}{m} x = 0
$$

Zavedeme stavov√© promƒõnn√©:

$$
x_1 = x \quad \text{(pozice)},\quad x_2 = \dot{x} \quad \text{(rychlost)}
$$

Pak:

$$
\dot{x}_1 = x_2 \\
\dot{x}_2 = -\frac{k}{m} x_1 - \frac{b}{m} x_2
$$

Matricovƒõ:

$$
\dot{\mathbf{x}} = \begin{bmatrix} 0 & 1 \\ -\frac{k}{m} & -\frac{b}{m} \end{bmatrix} \mathbf{x}
$$

---

## 2Ô∏è‚É£ Lagrangeova mechanika (energie)

Kinetick√° energie:  
\(T = \frac{1}{2} m \dot{x}^2\)

Potenci√°ln√≠ energie:  
\(V = \frac{1}{2} k x^2\)

Lagrangi√°n:

$$
\mathcal{L} = T - V = \frac{1}{2} m \dot{x}^2 - \frac{1}{2} k x^2
$$

Lagrangeova rovnice:

$$
\frac{d}{dt} \left( \frac{\partial \mathcal{L}}{\partial \dot{x}} \right) - \frac{\partial \mathcal{L}}{\partial x} = Q_{\text{diss}}
$$

Lev√° strana:

$$
\frac{d}{dt} (m \dot{x}) + k x = m \ddot{x} + k x
$$

Prav√° strana (dissipativn√≠ s√≠la):

$$
Q_{\text{diss}} = -b \dot{x}
$$

Celkov√° rovnice:

$$
m \ddot{x} + b \dot{x} + k x = 0
$$

Pak stejn√Ω p≈ôepis do stavov√©ho prostoru jako u Newtona.

---

## 3Ô∏è‚É£ Hamiltonova mechanika (hybnost)

Zavedeme hybnost:

$$
p = \frac{\partial \mathcal{L}}{\partial \dot{x}} = m \dot{x}
$$

Hamiltoni√°n (celkov√° energie):

$$
\mathcal{H}(x, p) = \frac{p^2}{2m} + \frac{1}{2} k x^2
$$

Hamiltonovy rovnice:

$$
\dot{x} = \frac{\partial \mathcal{H}}{\partial p} = \frac{p}{m}, \quad
\dot{p} = -\frac{\partial \mathcal{H}}{\partial x} = -k x
$$

P≈ôid√°n√≠ tlumen√≠ (mimo r√°mec konzervativn√≠ mechaniky):

$$
\dot{p} = -k x - \frac{b}{m} p
$$

Tedy:

$$
\dot{x} = \frac{p}{m}, \quad \dot{p} = -k x - \frac{b}{m} p
$$

Po substituci zpƒõt dostaneme stejnou rovnici pro \(\ddot{x}\).

---

## ‚úÖ Shrnut√≠: V≈°echny metody vedou ke stejn√© pohybov√© rovnici

$$
\ddot{x} + \frac{b}{m} \dot{x} + \frac{k}{m} x = 0
$$

A ta se n√°slednƒõ p≈ôepisuje do stavov√©ho prostoru:

$$
\dot{x}_1 = x_2 \\
\dot{x}_2 = -\frac{k}{m} x_1 - \frac{b}{m} x_2
$$

$$
\dot{\mathbf{x}} = A \mathbf{x}, \quad A = \begin{bmatrix} 0 & 1 \\ -\frac{k}{m} & -\frac{b}{m} \end{bmatrix}
$$

T≈ôi pohledy, stejn√Ω v√Ωsledek. üí°


## üìä Srovn√°n√≠ t≈ô√≠ p≈ô√≠stup≈Ø: Newton vs. Lagrange vs. Hamilton

| Vlastnost / Metoda           | Newtonova mechanika                  | Lagrangeova mechanika                      | Hamiltonova mechanika                      |
|-----------------------------|--------------------------------------|--------------------------------------------|--------------------------------------------|
| üîç Hlavn√≠ promƒõnn√©           | S√≠ly a zrychlen√≠                     | Energie (kinetick√° a potenci√°ln√≠)          | Energie a hybnost (stavov√Ω popis)          |
| üìê Formulace rovnic          | S√≠la = hmotnost √ó zrychlen√≠         | Euler-Lagrangeova rovnice                  | Hamiltonovy rovnice                        |
| üßÆ Poƒçet rovnic              | 2. ≈ô√°d diferenci√°ln√≠ rovnice        | 2. ≈ô√°d diferenci√°ln√≠ rovnice               | 1. ≈ô√°d ‚Äì dvojice rovnic (x, p)             |
| üß∞ Kdy pou≈æ√≠t                | Jednoduch√© mechanick√© syst√©my       | Komplexn√≠ syst√©my, obecn√© sou≈ôadnice       | Syst√©my v teorii ≈ô√≠zen√≠, kvantov√° fyzika   |
| üí° V√Ωhoda                    | Intuitivn√≠ a p≈ô√≠moƒçar√°               | Automaticky respektuje vazby               | Vhodn√© pro stavov√Ω popis, elegantn√≠ z√°pis  |
| ‚ö†Ô∏è Nev√Ωhoda                  | Obt√≠≈ænƒõ ≈ôe≈°√≠ vazby                   | Nen√≠ ide√°ln√≠ pro disipativn√≠ syst√©my       | Obt√≠≈ænƒõj≈°√≠ fyzik√°ln√≠ interpretace          |
| üîó Vazby a omezen√≠           | Nutno p≈ôidat ruƒçnƒõ                  | Vhodn√© pro syst√©my s omezen√≠mi             | Obt√≠≈ænƒõj≈°√≠ pr√°ce s omezen√≠mi               |
| üì¶ Stavov√Ω popis             | Nutno p≈ôev√©st na 1. ≈ô√°d              | Nutno p≈ôev√©st na 1. ≈ô√°d                    | P≈ôirozenƒõ ve stavov√©m tvaru                |
| üí≠ Interpretace              | ‚ÄûS√≠ly tlaƒç√≠ a tahaj√≠ tƒõlesa‚Äú        | ‚ÄûSyst√©m se pohybuje tak, aby minimalizoval akci‚Äú | ‚ÄûEnergie proud√≠ v syst√©mu‚Äú         |

---

### üìå Kdy pou≈æ√≠t kterou?

- **Newtonova mechanika**:
  - Nejjednodu≈°≈°√≠ pro p≈ô√≠m√© s√≠ly a mal√© poƒçty tƒõles.
  - Intuitivn√≠ (co t√°hne, to se pohybuje).
  - Vhodn√° pro zaƒç√°teƒçn√≠ky.

- **Lagrangeova mechanika**:
  - Vhodn√° pro slo≈æit√© syst√©my s v√≠ce stupni volnosti a vazbami.
  - Skvƒõl√© pro mechaniku robot≈Ø, kyvadel, kloubov√Ωch soustav.
  - Funguje v obecn√Ωch sou≈ôadnic√≠ch (nejen kartezi√°nsk√Ωch).

- **Hamiltonova mechanika**:
  - U≈æiteƒçn√° v teorii ≈ô√≠zen√≠, optimalizaci, kvantov√© fyzice.
  - Vede p≈ôirozenƒõ na **stavov√Ω model** (prvn√≠ ≈ô√°d).
  - Pou≈æ√≠v√° se v modern√≠ fyzice a analytick√© mechanice.

---

### üß† Shrnut√≠:

- V≈°echny t≈ôi p≈ô√≠stupy vedou (v klasick√Ωch syst√©mech) ke stejn√Ωm rovnic√≠m pohybu.
- Li≈°√≠ se **pohledem** na probl√©m a **matematick√Ωm n√°strojem**.
- Volba z√°vis√≠ na typu syst√©mu, po≈æadovan√©m v√Ωstupu a preferenc√≠ch (s√≠ly vs. energie vs. stav).


# üß± P≈ô√≠klad: Pru≈æina s tlumen√≠m

P≈ôedstav si:

- Hmotnost tƒõlesa: \( m \)
- Pru≈æina s tuhost√≠ \( k \)
- Tlumic√≠ prvek s odporem \( b \)
- \( x \) je v√Ωchylka tƒõlesa z rovnov√°≈æn√© polohy (kladnƒõ dol≈Ø)
- V syst√©mu nen√≠ ≈æ√°dn√° vnƒõj≈°√≠ s√≠la \( u(t) \) (zat√≠m)

---

## ‚öôÔ∏è Rovnice pohybu

S√≠la pru≈æiny:‚ÄÉ\( F_s = -k x \)  
S√≠la tlumiƒçe:‚ÄÉ\( F_d = -b \dot{x} \)

Celkov√° s√≠la podle Newtonova z√°kona:

$$
m \ddot{x} = -b \dot{x} - k x
\Rightarrow
\ddot{x} + \frac{b}{m} \dot{x} + \frac{k}{m} x = 0
$$

To je **diferenci√°ln√≠ rovnice druh√©ho ≈ô√°du**. P≈ôep√≠≈°eme ji do **syst√©mu dvou rovnic prvn√≠ho ≈ô√°du**, abychom z√≠skali stavov√Ω model.


---

## üîÅ Zavedeme stavov√© promƒõnn√©

Zavedeme:

- \( x_1 = x \quad \text{‚Äì pozice} \)
- \( x_2 = \dot{x} \quad \text{‚Äì rychlost} \)

Pak:

$$
\dot{x}_1 = x_2 \\
\dot{x}_2 = -\frac{k}{m} x_1 - \frac{b}{m} x_2
$$

---

## üßÆ Maticov√Ω z√°pis (stavov√Ω model)

\[
\begin{bmatrix}
\dot{x}_1 \\
\dot{x}_2
\end{bmatrix}
=
\underbrace{
\begin{bmatrix}
0 & 1 \\
-\frac{k}{m} & -\frac{b}{m}
\end{bmatrix}
}_{A}
\begin{bmatrix}
x_1 \\
x_2
\end{bmatrix}
+
\underbrace{
\begin{bmatrix}
0 \\
0
\end{bmatrix}
}_{B}
u
\]

---

## üéÆ P≈ôid√°n√≠ vstupn√≠ s√≠ly \( u(t) \)

Pokud p≈Øsob√≠ vnƒõj≈°√≠ s√≠la \( u(t) \), pak:

\[
m \ddot{x} = -b \dot{x} - k x + u(t)
\Rightarrow
\dot{x}_2 = -\frac{k}{m} x_1 - \frac{b}{m} x_2 + \frac{1}{m} u
\]

A matice \( B \) se zmƒõn√≠ na:

\[
B =
\begin{bmatrix}
0 \\
\frac{1}{m}
\end{bmatrix}
\]

---

## üì§ V√Ωstup (nap≈ô. mƒõ≈ô√≠me pouze pozici)

\[
y =
\underbrace{
\begin{bmatrix}
1 & 0
\end{bmatrix}
}_{C}
\begin{bmatrix}
x_1 \\
x_2
\end{bmatrix}
+ \underbrace{0}_{D} \cdot u
\]

---

## ‚úÖ Fin√°ln√≠ stavov√Ω model pru≈æiny s tlumen√≠m

\[
\dot{x} =
\begin{bmatrix}
0 & 1 \\
-\frac{k}{m} & -\frac{b}{m}
\end{bmatrix}
x +
\begin{bmatrix}
0 \\
\frac{1}{m}
\end{bmatrix}
u
\]

\[
y =
\begin{bmatrix}
1 & 0
\end{bmatrix}
x + 0 \cdot u
\]


```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
```

```python
# Parametry syst√©mu
m = 1.0  # hmotnost [kg]
k = 2.0  # tuhost pru≈æiny [N/m]
b = 1.0   # koeficient tlumen√≠ [N¬∑s/m]

# Pravoƒçlen diferenci√°ln√≠ rovnice (stavov√Ω model)
def spring_system(state, t):
    x1, x2 = state  # x1 = pozice, x2 = rychlost
    dx1dt = x2
    dx2dt = (-k * x1 - b * x2) / m
    return [dx1dt, dx2dt]

# Poƒç√°teƒçn√≠ podm√≠nky: x = 1 m, v = 0 m/s
initial_state = [1.0, 0.0]

# ƒåasov√° osa pro simulaci (nap≈ô. 0 a≈æ 10 s)
t = np.linspace(0, 4, 400)

# Integrace syst√©mu
solution = odeint(spring_system, initial_state, t)

# V√Ωstup: solution je n x 2 pole (pozice a rychlost)
print("V√Ωstup shape:", solution.shape)  # (1000, 2)
```

```python
# üîß P≈ô√≠prava animace
%matplotlib notebook
fig_of_plot, axes = plt.subplots(2, 1, constrained_layout=True, figsize=(7, 10))

# === PRVN√ç GRAF ‚Äì Pohyb pru≈æiny (2D pohled z boku) ===
axes[0].set_xlim((-0.5, 0.5))
axes[0].set_ylim((-1.2, 1.2))
axes[0].set_title('Spring movement')
axes[0].set_aspect('equal')
axes[0].grid()

# === DRUH√ù GRAF ‚Äì Hodnoty v ƒçase ===
axes[1].set_xlim((0, len(solution)))
axes[1].set_ylim((-1.5, 1.5))
axes[1].set_title('Spring attributes')
axes[1].grid()

# === Pr√°zdn√© ƒç√°ry pro animaci ===
mass_point, = axes[0].plot([], [], 'bo', markersize=10)
spring_line, = axes[0].plot([], [], 'b-', linewidth=2)
position_line, = axes[1].plot([], [], 'g-', label='position [m]')
velocity_line, = axes[1].plot([], [], 'r-', label='velocity [m/s]')
axes[1].legend()

# === Aktualizaƒçn√≠ funkce ===
def update_plot(i):
    if i < 2:
        return

    # Data a≈æ do aktu√°ln√≠ho kroku
    state_until_now = solution[:i]
    y_vals = state_until_now[:, 0]
    v_vals = state_until_now[:, 1]

    current_y = y_vals[-1]

    # Pru≈æina a bod
    mass_point.set_data([0], [-current_y])
    spring_line.set_data([0, 0], [0, -current_y])

    # Grafy pr≈Øbƒõhu
    position_line.set_data(np.arange(i), y_vals)
    velocity_line.set_data(np.arange(i), v_vals)

# === Spu≈°tƒõn√≠ animace ===
ani = animation.FuncAnimation(fig_of_plot, update_plot, frames=len(solution), interval=10, repeat=False)

HTML(ani.to_jshtml())
```

```python
# Parametry syst√©mu
m = 1.0  # hmotnost [kg]
k = 2.0  # tuhost pru≈æiny [N/m]
b = 1.0  # koeficient tlumen√≠ [N¬∑s/m]

# Stavov√Ω model odvozen√Ω z Lagrangeovy rovnice
def lagrangian_spring_system(state, t):
    x, v = state  # x = pozice, v = rychlost
    dxdt = v
    dvdt = (-k * x - b * v) / m
    return [dxdt, dvdt]

# Poƒç√°teƒçn√≠ podm√≠nky: x = 1 m, v = 0 m/s
initial_state = [1.0, 0.0]

# ƒåasov√° osa pro simulaci
t = np.linspace(0, 4, 400)

# Integrace syst√©mu
solution = odeint(lagrangian_spring_system, initial_state, t)

# V√Ωstup: solution je n x 2 pole (pozice a rychlost)
print("V√Ωstup shape:", solution.shape)  # (1000, 2)
```

```python
# üîß P≈ô√≠prava animace
%matplotlib notebook
fig_of_plot, axes = plt.subplots(2, 1, constrained_layout=True, figsize=(7, 10))

# === PRVN√ç GRAF ‚Äì Pohyb pru≈æiny (2D pohled z boku) ===
axes[0].set_xlim((-0.5, 0.5))
axes[0].set_ylim((-1.2, 1.2))
axes[0].set_title('Spring movement')
axes[0].set_aspect('equal')
axes[0].grid()

# === DRUH√ù GRAF ‚Äì Hodnoty v ƒçase ===
axes[1].set_xlim((0, len(solution)))
axes[1].set_ylim((-1.5, 1.5))
axes[1].set_title('Spring attributes')
axes[1].grid()

# === Pr√°zdn√© ƒç√°ry pro animaci ===
mass_point, = axes[0].plot([], [], 'bo', markersize=10)
spring_line, = axes[0].plot([], [], 'b-', linewidth=2)
position_line, = axes[1].plot([], [], 'g-', label='position [m]')
velocity_line, = axes[1].plot([], [], 'r-', label='velocity [m/s]')
axes[1].legend()

# === Aktualizaƒçn√≠ funkce ===
def update_plot(i):
    if i < 2:
        return

    # Data a≈æ do aktu√°ln√≠ho kroku
    state_until_now = solution[:i]
    y_vals = state_until_now[:, 0]
    v_vals = state_until_now[:, 1]

    current_y = y_vals[-1]

    # Pru≈æina a bod
    mass_point.set_data([0], [-current_y])
    spring_line.set_data([0, 0], [0, -current_y])

    # Grafy pr≈Øbƒõhu
    position_line.set_data(np.arange(i), y_vals)
    velocity_line.set_data(np.arange(i), v_vals)

# === Spu≈°tƒõn√≠ animace ===
ani = animation.FuncAnimation(fig_of_plot, update_plot, frames=len(solution), interval=10, repeat=False)

HTML(ani.to_jshtml())
```

```python
# Parametry syst√©mu
m = 1.0  # hmotnost [kg]
k = 2.0  # tuhost [N/m]
b = 0.5  # tlumen√≠ [N¬∑s/m]

# Hamiltonovsk√Ω model
def hamiltonian_spring_system(state, t):
    x, p = state
    dxdt = p / m
    dpdt = -k * x - (b / m) * p
    return [dxdt, dpdt]

# Poƒç√°teƒçn√≠ podm√≠nky: x = 1 m, p = 0 kg¬∑m/s
initial_state = [1.0, 0.0]

# ƒåasov√° osa
t = np.linspace(0, 4, 400)

# Integrace
solution = odeint(hamiltonian_spring_system, initial_state, t)
```
